```python
from matplotlib import pyplot as plt
import numpy as np,pandas as pd
%matplotlib inline
```

```python
data = [1,4,34,54,78,12,89,13,90,120,210]
len(data)
```

```python
iqr1 , iqr3 = np.quantile(data , [.25 , .75])
median = np.median(data)
lower_lim = (iqr1-(1.5*(iqr3-iqr1)))
upper_lim = (iqr3+(1.5*(iqr3-iqr1)))
```

```python
for x in data:
    if x> upper_lim or x<lower_lim:
        print("{} is an outlier".format(x))
```

```python
plt.figure(figsize=(10,10))
plt.ylabel("Box plots")
plt.xlabel("Values of the data set")
plt.title("Box plot for the data in the data variable.")
plt.boxplot(data , vert=False , patch_artist=True , notch=True);
```
